```python
import pandas as pd
```

```python
df=pd.read_csv("D:\project1\Product-Sales-Region.csv")
```

```python
df
```

```python
df.isnull().sum()
```

```python
import matplotlib.pyplot as plt
```

```python
region_sales = df.groupby("Region")["TotalPrice"].sum()
plt.pie(region_sales, labels=region_sales.index, autopct='%1.1f%%')
```

```python
StoreLocation_sales = df.groupby("StoreLocation")["TotalPrice"].sum()

plt.bar(StoreLocation_sales.index,
        StoreLocation_sales.values,
        color='green',
        edgecolor='black',
        linewidth=2)

plt.title("Total Sales by Store Location")
plt.xlabel("Store Location")
plt.ylabel("Total Sales")
plt.show()
```

```python
region_sales = df.groupby("Region")["TotalPrice"].sum()

plt.figure(figsize=(8,5))
plt.bar(region_sales.index,
        region_sales.values,
        color="orange",
        edgecolor="black")

plt.title("Total Sales by Region")
plt.xlabel("Region")
plt.ylabel("Total Sales")
plt.show()
```

```python
import matplotlib.pyplot as plt

top_products = df.groupby("Product")["TotalPrice"].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(10,5))
plt.bar(top_products.index,
        top_products.values,
        color="purple",
        edgecolor="black")

plt.title("Top 10 Products by Sales")
plt.xlabel("Product")
plt.ylabel("Total Sales")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
